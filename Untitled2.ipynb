## Model Testing and Training

```python
#Split the data
X = df.iloc[:,1:df.shape[1]].values
Y = df.iloc[:,0].values
```

```python
print(Counter(Y))
rus = RandomOverSampler(random_state = 42)
X_over, Y_over = rus.fit_resample(X,Y)
print(Counter(Y_over))
```

```python
#Split the data into 75% training and 25% testing
X_train, X_test, Y_train, Y_test = train_test_split(X_over,Y_over, test_size = 0.25, random_state = 42)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
#Random Forest Classifier
model = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 42 )
```

```python
model.fit(X_train, Y_train)
```

```python
#Get accuracy on training dataset
model.score(X_train,Y_train)
```

```python
#Confusion matrix and accuracy score for the model on test data
cm = confusion_matrix(Y_test,model.predict(X_test))
TN= cm[0][0]
TP = cm[1][1]
FN = cm[1][0]
FP = cm[0][1]
print(cm)
print('model testing accuracy = {}'.format((TP+TN)/(TP+TN+FP+FN) ))
```

```python
print(classification_report(Y_test, model.predict(X_test)))
```

```python
print(accuracy_score(Y_test, model.predict(X_test)))
```
